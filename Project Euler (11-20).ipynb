# 12.) Highly Divisible Triangular Number

The sequence of triangle numbers is generated by adding the natural numbers. So the $7^{th}$ triangle number would be $1 + 2 + 3 + 4 + 5 + 6 + 7 = 28$. The first ten terms would be:

$$1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ... $$

Let us list the factors of the first seven triangle numbers:

$ 1: 1 \\ $
$ 3: 1,3 \\ $
$ 6: 1,2,3,6 \\ $
$ 10: 1,2,5,10 \\ $
$ 15: 1,3,5,15 \\ $
$ 21: 1,3,7,21 \\ $
$ 28: 1,2,4,7,14,28 \\ $

We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?

```python
# Triangular number formula: n*(n+1)/2.

# Importing relevant libraries
from math import sqrt, ceil

# STEP 1: DIVISORS

# Function to determine if a number is square
def isSquare(n):
    return (int(sqrt(n))**2 == n)

# Function to determine the number of divisors
def divisorNumber(n):
    
    # No divisors initally
    count = 0
    
    # Square number condition
    square = isSquare(n)
    
    # Looping until the square-root
    for i in range(1, ceil(sqrt(n))):
        if (n%i == 0):
            count +=1
    
    # Number of divisors depend on square number
    if (square == True):
        return (2*count - 1)
    else:
        return (2*count)

# STEP 2: SOLUTION
    
# Initial state of the problem is unsolved.
solved = False

# Looping until the problem is solved
i = 1
while (not solved):
    
    # Triangle number
    triangle = i*(i+1)/2
    
    # Number of divisors
    div = divisorNumber(triangle)
    
    # If it's above 500...
    if (div > 500):
        
        # ...problem is solved.
        result = triangle
        solved = True
    
    # Otherwise...
    else:
        # ...try the next number
        i += 1

print(result) #76576500
```

# 14.) Longest Collatz Sequence

$ n → n/2 $ (n is even) $\\$
$ n → 3n + 1 $ (n is odd)

Using the rule above and starting with 13, we generate the following sequence:

$ 13 → 40 → 20 → 10 → 5 → 16 → 8 → 4 → 2 → 1 $

It can be seen that this sequence (starting at 13 and finishing at 1) contains 10 terms. Although it has not been proved yet (Collatz Problem), it is thought that all starting numbers finish at 1.

Which starting number, under one million, produces the longest chain?

NOTE: Once the chain starts the terms are allowed to go above one million.

```python
# Initial dictionary (nummber: chain length)
sequence = {}

# Recursive collatz function
def collatz(n):
    chain = 1
    if (n==1):
        return chain
    else:
        if (n%2 == 0):
            n /= 2
        else:
            n = 3*n+1
        return (1 + collatz(n))

# Looping until 1 million
for i in range(1, 1000000):
    sequence[i] = collatz(i)

# Get the number with maximum collatz sequence
result = max(sequence, key=sequence.get)

print(result) #837799
```

# 15.) Lattice Paths

Starting in the top left corner of a 2×2 grid, and only being able to move to the right and down, there are exactly 6 routes to the bottom right corner.

<img src="https://projecteuler.net/project/images/p015.png">

How many such routes are there through a 20×20 grid?

```python
# Grid size
g = 20

# There are always g right and g down move.
# Formula is: (2*g)! / (g! * g!)

# Factorial function
def factorial(n):
    if (n==0):
        return 1
    elif (n==1):
        return 1
    else:
        return n*factorial(n-1)
    
# Route function
def routeNumber(grid):
    formula = factorial(2*grid) / (factorial(grid)**2)
    return formula

# Number of routes for a g*g grid
result = routeNumber(g)

print(result) #137846528820
```

If the numbers 1 to 5 are written out in words: one, two, three, four, five, then there are 3 + 3 + 5 + 4 + 4 = 19 letters used in total.

If all the numbers from 1 to 1000 (one thousand) inclusive were written out in words, how many letters would be used?


**NOTE**: Do not count spaces or hyphens. For example, 342 (three hundred and forty-two) contains 23 letters and 115 (one hundred and fifteen) contains 20 letters. The use of "and" when writing out numbers is in compliance with British usage.

```python
# STEP 1: Below 100

# Until twenty
single = [0,
          len("one"),
          len("two"),
          len("three"),
          len("four"),
          len("five"),
          len("six"),
          len("seven"),
          len("eight"),
          len("nine"),
          len("ten"),
          len("eleven"),
          len("twelve"),
          len("thirteen"),
          len("fourteen"),
          len("fifteen"),
          len("sixteen"),
          len("seventeen"),
          len("eighteen"),
          len("nineteen")]

# Tenths
tenths = [0,
          0,
          len("twenty"),
          len("thirty"),
          len("forty"),
          len("fifty"),
          len("sixty"),
          len("seventy"),
          len("eighty"),
          len("ninety")]

# Function: number of letters for a number <100
def lettersBelow100(n):
    if (n < 20):
        return(single[n])
    else:
        remainder = n%10
        tenth = int((n - remainder)/10)
        return(tenths[tenth]+single[remainder])
    
# Function: number of letters for a number >= 100   
def lettersAbove100(n):
    h = n - (n%100)
    hundreth = int(h/100)
    remainder = n-h
    return(single[hundreth]+len("hundredand")+lettersBelow100(remainder))

# Function to combine these
def noOfLetters(n):
    if(n==1000):
        return(len("onethousand"))
    else:
        if(n < 100):
            return(lettersBelow100(n))
        else:
            return(lettersAbove100(n))

# Initial result is unknown (0).
result = 0

# Looping until one thousand
for i in range(1,1001):
    result += noOfLetters(i)

print(result) #21124
```

# 19.) Counting Sundays

You are given the following information, but you may prefer to do some research for yourself.

- 1 Jan 1900 was a Monday.
- Thirty days has September,
  April, June and November.
  All the rest have thirty-one,
  Saving February alone,
  Which has twenty-eight, rain or shine.
  And on leap years, twenty-nine.
- A leap year occurs on any year evenly divisible by 4, but not on a century unless it is divisible by 400.

How many Sundays fell on the first of the month during the twentieth century (1 Jan 1901 to 31 Dec 2000)?

```python
# Importing relevant libraries
import datetime
from dateutil.relativedelta import relativedelta

# Starting and ending dates
start = datetime.date(1901, 1, 1)
end = datetime.date(2000, 12, 31)
delta = relativedelta(months=1)

# Initial result is unknown (0).
result = 0

# Looping through the dates
while (start <= end):
    if (start.strftime('%A') == "Sunday"):
        result += 1
    start += delta

print(result) #171
```

# 20.) Factorial Digit Sum

n! means n × (n − 1) × ... × 3 × 2 × 1

For example, 10! = 10 × 9 × ... × 3 × 2 × 1 = 3628800,
and the sum of the digits in the number 10! is 3 + 6 + 2 + 8 + 8 + 0 + 0 = 27.

Find the sum of the digits in the number 100!

```python
# Value in question
n = 100

# Factorial function
def factorial(n):
    if (n==0):
        return 1
    elif (n==1):
        return 1
    else:
        return n*factorial(n-1)
    
# Digit sum function
def digitSum(n):
    number = str(n)
    sum_value = 0
    for digit in number:
        sum_value += int(digit)
    return sum_value

result = digitSum(factorial(n))

print(result) #648
```
